```python
f=open('result2.txt', 'r')
cnt = 0
filt = []
layer = []
drop = []
acc = []
loss = []
iou = []

for line in f.readlines():
    if cnt % 10 == 0:
        filt.append(int(line[:-1]))
    elif cnt % 10 == 1:
        layer.append(int(line[:-1]))
    elif cnt % 10 == 2:
        drop.append(int(line[:-1]))
    elif cnt % 10 == 3:
        acc.append(float(line[:-1]))
    elif cnt % 10 == 4:
        loss.append(float(line[:-1]))
    elif cnt % 10 == 5:
        iou.append(float(line[:-1]))
    cnt += 1
f.close()
```
